```python
#Importing the necessary libraries
import pandas as pd
import numpy as np
```

```python
#Loading the dataset    
df = pd.read_csv("datasets/AfricaCupofNationsMatches.csv")
#Displaying the first few rows of the dataset
df.head()
```

```python
#Selecting the columns of interest
df[['HomeTeam', 'AwayTeam', 'HomeTeamGoals', 'AwayTeamGoals']]
```

```python
#Select wrows where the home or away team is 'Egypt'
df[(df['HomeTeam'] == 'Egypt') | (df['AwayTeam'] == 'Egypt')]
```

```python
#The shape of the DataFrame
df.shape
```

```python
#Adding a column to the end of the end of the DataFrame
df['TotalGoals'] = df['HomeTeamGoals'] + df['AwayTeamGoals']
#Displaying the first few rows of the modified DataFrame
df.head()
```

```python
#Adding 2 rows to the end of the DataFrame 
new_rows = pd.DataFrame([
    {'HomeTeam': 'Egypt', 'AwayTeam': 'Nigeria', 'HomeTeamGoals': 2, 'AwayTeamGoals': 1, 'TotalGoals': 3},
    {'HomeTeam': 'Nigeria', 'AwayTeam': 'Egypt', 'HomeTeamGoals': 1, 'AwayTeamGoals': 2, 'TotalGoals': 3}
])
df = pd.concat([df, new_rows], ignore_index=True)
df.tail(2)
```

```python
#a Pandas program to change the country ‘Uganda’ to ‘China’ in the ‘AwayTeam’ column of the DataFrame
df['AwayTeam'] = df['AwayTeam'].replace('Uganda', 'China')
#Confirming the change
df[df['AwayTeam'] == 'China']
```

```python
#14.a Pandas program to reset index in your DataFrame.
df.reset_index(drop=True, inplace=True)
#Confirming the reset index
df.head()
```

```python
# Pandas program to check whether the ‘Stadium’ column is present in your DataFrame or not
'Stadium' in df.columns
```

```python
#Pandas program to convert the datatype of the ‘AwayTeamGoals’. (int to float)
df['AwayTeamGoals'] = df['AwayTeamGoals'].astype(float)
#Confirming the datatype change
df['AwayTeamGoals'].dtype
```

```python
#Pandas program to remove the last 10 rows from your DataFrame
df = df[:-10]
#Confirming the removal of the last 10 rows
df.tail(10)
```

```python
#Pandas program to iterate over rows in your DataFrame.
for index, row in df.iterrows():
    print(f"Match {index + 1}: {row['HomeTeam']} vs {row['AwayTeam']} - {row['HomeTeamGoals']}:{row['AwayTeamGoals']}")
```

```python
#Pandas program to change the order of your DataFrame columns
df = df[['AwayTeam', 'HomeTeam', 'AwayTeamGoals', 'HomeTeamGoals', 'TotalGoals']]
#Confirming the change in column order
df.head()
```

```python
#Pandas program to delete DataFrame row(s) whose value is 0 in the ‘HomeTeamGoals’ columns
df = df[df['HomeTeamGoals'] != 0]
#Confirming the deletion of rows with 0 in 'HomeTeamGoals'
df[df['HomeTeamGoals'] == 0]
```
